```python
# step 4: 有切換三區段點位的list了, 計算三區段分數, 
# step 4-1: partselect function
"""
partselect function
篩選出 各區段有啟用值和無啟用值的 list

--> 一個單位一個單位計算 residual
-->合併成 一 residual dataframe
-->押出段處理function
"""
"""
partselect func
輸入: opdf(存允差) ,點位清單, 部位 
輸出: 點位keyword, 點位tag (差值處理寫在 healthvalue裡)
 """

def partselect(machinedict, machine, part):

    if machine=='HHR3':
        if part =='押出機':
            yachu_alllist = machinedict[part][0] # 押出機全部點位
            # 篩選 三種: 有啟用值成對，無啟用值成對，單獨點位      
            yachu_alone_kw = ['押出機電流實際值(%)','真空壓力實際值(mmHg)']
            yachu_pair_kw = ['供料速度','押出機速度',]
            yachu_pair_en_kw = [item.split('溫度啟用')[0] for item in yachu_alllist if '溫度啟用' in item]
            return yachu_alllist, yachu_pair_en_kw, yachu_pair_kw, yachu_alone_kw


        elif part =='三本ROLL':
            roll_alllist = machinedict[part][0]

            roll_pair_kw = ['上輪速度', '中輪速度', '下輪速度']
            roll_pair_en_kw = ['上輪溫度', '中輪溫度','下輪溫度']
            roll_alone_kw = []
            return roll_alllist, roll_pair_en_kw, roll_pair_kw, roll_alone_kw

        elif part=='引取機':
            inch_alllist = machinedict[part][0]
            inch_pair_kw = ['引取速度']
            inch_pair_en_kw = []
            inch_alone_kw = []
            return inch_alllist, inch_pair_en_kw, inch_pair_kw, inch_alone_kw
```

## 押出段 三部分 function 處理完成, 接著要和在一起
#### 正式 function 如下

```python
# step 4-2 : 根據 區段篩選條件去建立 healthvalue func
"""
輸入: 全部點位dataframe(chdf), op資料(op dataframe) -->允差從這裡拿
        , machinepart(區段對應 dict) --> 點位從這裡拿 ,區段(''HHR3)
輸出:輸出 residual dataframe, df_final(允差, 扣分值, 分數), final_score (彙總總分)
"""
# 三段分別為 yachu_pair_en_kw, yachu_pair_kw, yachu_alone_kw
def yachu_hv1(df_, opdf, machinedict):
    total, en_pair, pair, alone = partselect(machine_dict, 'HHR3', '押出機')
    # 目標生成三個 處理 1. pair_en
    df_res1 = pd.DataFrame()
    df_Final1=pd.DataFrame(index=["residual", "f_i1", "F_i"])
    # 篩選有開機點位
    for i, col in enumerate(en_pair):
            
        if col =='模具D1':
            usecol = ['模具D1溫度啟用','模具D1實際溫度值（℃）','模具D1設定溫度值（℃）']
        else:
            usecol = [item for item in df_.columns.values if col in item]
        usecol.remove(col+'溫度啟用') # 篩選出有啟用, 的實際和設定值

        # 押出段有啟用配對值的 df_res1
        if df_[usecol[0]].dtypes == float and df_[usecol[1]].dtypes == float:
            df_res1[(col+'溫度')] = abs(df_[usecol[0]] - df_[usecol[1]]) 
        else:
            df_res1[(col+'溫度')] = 0

        # 處理允差 limit 目前有允差 
        limit = opdf[opdf['量測值名稱']==(col+'溫度啟用')]['標準值'].reset_index(drop = True)[0]
        if type(limit) == str:
            if df_['模具D13溫度啟用'].dtypes == float:
                if df_['模具D13溫度啟用'].mean()==1.0:
                    limit = 2
                else:
                    limit = 20
            else:
                limit = 20
        meannow = df_res1[(col+'溫度')].mean()
        if df_[(col+'溫度啟用')].dtypes == float:
            if df_[(col+'溫度啟用')].mean() ==1.0: # 代表那個點位正常生產
                # 計算 f_i1 值
                f_i1 = 20*(meannow/limit)
            else:
                f_i1 = 100 # 非正常生產, 扣分100
        else:
            f_i1 = 100
        
        temp = 100 - f_i1
        if temp < 0:
            temp = 0

        # 有各別允差還有 點位分數
        df_Final1[(col+'溫度')] = [limit, f_i1, temp]
            

    return df_res1, df_Final1
```

```python
def yachu_hv2(df_, opdf, machinedict):
    total, en_pair, pair, alone = partselect(machine_dict, 'HHR3', '押出機')
    # 目標生成三個 處理 2. pair
    df_res2 = pd.DataFrame()
    df_Final2=pd.DataFrame(index=["residual", "f_i1", "F_i"])
    for i, col in enumerate(pair):
        usecol = [item for item in df_.columns.values if col in item]
        # 押出段沒啟用值的配對值res2
        if df_[usecol[0]].dtypes == float and df_[usecol[1]].dtypes == float:
            df_res2[(col)] = abs(df_[usecol[0]] - df_[usecol[1]]) 
        else:
            df_res2[(col)] = np.zeros(len(df_[usecol[0]]))
        
        # 計算允差和扣分
        limit = opdf[opdf['量測值名稱']==(col+'實際值(RPM)')]['標準值'].reset_index(drop = True)[0]
        meannow = df_res2[col].mean()

        if df_[usecol[0]].dtypes == float and df_[usecol[1]].dtypes == float:
            if df_[usecol[0]].mean() ==0 and df_[usecol[1]].mean() ==0:
                f_i1 = 100 # 非正常生產, 扣分100
            else:
                f_i1 = 20*(meannow/limit) # 正常生產扣分狀態
        else:
            f_i1 = 100
        
        temp = 100 - f_i1
        if temp < 0:
            temp = 0
        # 有各別允差還有 點位分數
        df_Final2[(col)] = [limit, f_i1, temp]
        df_res2.index = df_.index

    return df_res2, df_Final2
```

```python
def yachu_hv3(df_, opdf, machinedict):
    total, en_pair, pair, alone = partselect(machine_dict, 'HHR3', '押出機')
    # 目標生成三個 處理 3. alone
    df_res3 = pd.DataFrame()
    df_Final3=pd.DataFrame(index=["residual", "f_i1", "F_i"])
    for i, col in enumerate(alone):
        # 計算允差和扣分
        if col =='真空壓力實際值(mmHg)':
            if df_[col].dtypes ==float:
                meannow = df_[col].mean()
                limit = '>=680'
                if meannow >=680:
                    f_i1 = 0  
                else:
                    f_i1 = 100 # 真空值小於 680, 扣分給100
                df_res3[col] = df_[col]
            else:
                f_i1 = 100
                df_res3[col] = np.zeros(len(df_[col]))

        elif col =='押出機電流實際值(%)':
            if df_[col].dtypes ==float:
                df_res3[col] = abs(df_[col]-50)
                limit = 25
                meannow = df_res3[col].mean()
                f_i1 = 20*(meannow/limit)
            else:
                limit = 25
                f_i1 = 100
                df_res3[col] = np.zeros(len(df_[col]))

        temp = 100 - f_i1
        if temp < 0:
            temp = 0
        df_Final3[(col)] = [limit, f_i1, temp]
        df_res3.index = df_.index

    return df_res3, df_Final3
```

### (完成)合併 押出三段 寫一個 function, 回傳 合併後 residual df, 扣分 df, 總分

```python
def yachu_hv(df_, opdf, machinedict):
    # 回傳 三段yachu_pair_en_kw, yachu_pair_kw, yachu_alone_kw, 計算之 residual df 和 有扣分值得df_final
    df_res_1,df_Final_1= yachu_hv1(df_, opdf, machinedict)
    df_res_2,df_Final_2= yachu_hv2(df_, opdf, machinedict)
    df_res_3,df_Final_3= yachu_hv3(df_, opdf, machinedict)

    # 合併
    df_res = pd.concat([df_res_1, df_res_2 ,df_res_3], axis = 1)
    df_Final = pd.concat([df_Final_1, df_Final_2, df_Final_3], axis = 1)

    diff = ['模具D10溫度', '模具D11溫度', '模具D12溫度','模具D13溫度']

    # act_num 判斷 開9 or 13
    if df_Final.loc['F_i',diff].sum() == 0.0:
        df_Final_temp = df_Final[[elm for elm in df_Final.columns.values if elm not in diff]]
        act_num = 9
    else:
        df_Final_temp = df_Final
        act_num = 13

    if df_Final_temp.loc['F_i'].min() < 80:
        Final_Score = df_Final_temp.loc['F_i'].min()

    elif df_Final_temp.loc['F_i'].min() >= 80:
        Final_Score = df_Final_temp.loc['F_i'].mean()

    return df_res, df_Final, Final_Score , act_num   
```

### (完成)寫 三本 ROLL 計算 function (上、中、下輪、引取 速度的實際和設定，若非正常生產，會回傳 字典)

```python
# 三段分別為 rool_pair_en_kw, roll_pair_kw, roll_alone_kw
def roll_hv1(df_, opdf, machinedict):
    total, en_pair, pair, alone = partselect(machine_dict, 'HHR3', '三本ROLL')
    # 目標生成三個 處理 1. pair_en
    df_res1 = pd.DataFrame()
    df_Final1=pd.DataFrame(index=["residual", "f_i1", "F_i"])
    # 篩選有開機點位
    for i, col in enumerate(en_pair):
        usecol = [item for item in df_.columns.values if col in item]
        usecol.remove(col+'啟用') # 篩選出有啟用, 的實際和設定值
        # 押出段有啟用配對值的 df_res1
        if df_[usecol[0]].dtypes == float and df_[usecol[1]].dtypes == float:
            df_res1[(col)] = abs(df_[usecol[0]] - df_[usecol[1]]) 
        else:
            df_res1[(col)] = np.zeros(len(df_[usecol[0]]))

        
        # 處理允差 limit 目前有允差 
        limit = opdf[opdf['量測值名稱']==(col+'啟用')]['標準值'].reset_index(drop = True)[0]
        meannow = df_res1[(col)].mean()
        if df_[col+'啟用'].dtypes == float:
            if df_[(col+'啟用')].mean() ==1.0: # 代表那個點位正常生產
                # 計算 f_i1 值
                f_i1 = 20*(meannow/limit)
            else:
                f_i1 = 100 # 非正常生產, 扣分100
        else:
            f_i1 = 100
        temp = 100 - f_i1
        if temp < 0:
            temp = 0
        # 有各別允差還有 點位分數
        df_Final1[(col)] = [limit, f_i1, temp]
        df_res1.index = df_.index

    return df_res1, df_Final1
```

```python
# 三段分別為 rool_pair_en_kw, roll_pair_kw, roll_alone_kw
def roll_hv2(df_, opdf, machinedict):
    total, en_pair, pair, alone = partselect(machine_dict, 'HHR3', '三本ROLL')
    
    df_res2 = pd.DataFrame()
    df_Final2=pd.DataFrame(index=["residual", "f_i1", "F_i"])
    # 篩選有開機點位
    for i, col in enumerate(pair):
        usecol = [item for item in total if col in item]

        limit = opdf[opdf['量測值名稱']==(col+'實際值(M/min)')]['標準值'].reset_index(drop = True)[0]

        # 若非正常生產, 會回傳字典
        if df_[usecol[0]].dtypes == object and df_[usecol[1]].dtypes == object:
            df_res2[col] = [0] * len(df_) # 非正常生產, 殘差值給 0
            df_res2.index = df_.index
            f_i1 = 100 # 扣分給100
        
        else: # 正常生產狀態
            # 押出段沒啟用值的配對值res2
            df_res2[col] = abs(df_[usecol[0]] - df_[usecol[1]])
            # 計算允差和扣分
            meannow = df_res2[col].mean()
            f_i1 = 20*(meannow/limit)
        temp = 100 - f_i1
        if temp < 0:
            temp = 0

        # 有各別允差還有 點位分數
        df_Final2[(col)] = [limit, f_i1, temp]

    return df_res2, df_Final2
```

## (完成)合併三本ROLL段

```python
def roll_hv(df_, opdf, machinedict):
    # 回傳 三段yachu_pair_en_kw, yachu_pair_kw, yachu_alone_kw, 計算之 residual df 和 有扣分值得df_final
    df_res_1,df_Final_1= roll_hv1(df_, opdf, machinedict)
    df_res_2,df_Final_2= roll_hv2(df_, opdf, machinedict)

    # 合併
    df_res = pd.concat([df_res_1, df_res_2], axis = 1)
    df_Final = pd.concat([df_Final_1, df_Final_2], axis = 1)

    if df_Final.loc['F_i'].min() < 80:
        Final_Score = df_Final.loc['F_i'].min()
    elif df_Final.loc['F_i'].min() >= 80:
        Final_Score = df_Final.loc['F_i'].mean()

    return df_res, df_Final, Final_Score    
```
